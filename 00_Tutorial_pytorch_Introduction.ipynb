Preprocessing the image</br>
scaling 256*256</br>
crop image</br>
normalize its RGB


```python
from torchvision import transforms
```

```python
preprocess=transforms.Compose({transforms.Resize(256),transforms.CenterCrop(224),transforms.ToTensor(),transforms.Normalize(mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225])})
```

```python
from PIL import Image
```

```python
img=Image.open("dog.jpg")
```

```python
img
```

```python
img_t=preprocess(img)
```

```python
img_t
```

```python
batch_t=torch.unsqueeze(img_t,0)
```

The process of running a trained model on new data is called interface. Inorder to do interface, we need to put the network in eval mode.

```python
resent.eval()
```

```python
out=resent(batch_t)
```

```python
out
```

```python
with open ("imagenet_classes.txt") as f:
  labels=[line.strip() for line in f.readlines()]
```

```python
_, index=torch.max(out,1)
```

```python
percentage=torch.nn.functional.softmax(out,dim=1)[0]*100
```

```python
labels[index[0]],percentage[index[0]].item()
```

```python
_, indices=torch.sort(out,descending=True)
[(labels[idx],percentage[idx].item()) for idx in indices[0][0:5]]
```
